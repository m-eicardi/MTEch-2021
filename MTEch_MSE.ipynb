*1.3.3 Existe correlação entre o número de vendas e o número de unidades?*

```python
#importando e plotando gráfico de correlação com seaborn
import seaborn as sns

sns.set_theme(style="darkgrid")
sns.lmplot(x='Units', y='Sales', data=df) #linha de tendencia
```
