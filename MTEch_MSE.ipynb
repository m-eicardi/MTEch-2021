```python
netflix_df.loc[index, 'duration']
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
spotify = pd.read_csv('top10s (1).csv', encoding='ISO-8859-1')

spotify
```

```python
#eliminar coluna sem nome

spotify.drop(columns=['Unnamed: 0'], inplace=True)

spotify
```

```python
spotify.columns
```

```python
#renomear colunas
spotify.columns = ['título', 'artista', 'gênero', 'ano', 'bpm', 'energia',
       'dançabilidade', 'volume ', 'liveness', 'positividade', 'duração',
       'acústica', 'oralidade', 'popularidade']

spotify
```

```python
#descrições estatísticas transposta
#média, desviop, min, q1, q2(mediana),q3 e max
spotify.describe().T
```

```python
#gráfico 
sns.boxplot(y=spotify['ano'])

plt.show()
```

```python
sns.boxplot(y=spotify['acústica'])

plt.show()
```

```python
Q = spotify['acústica'].quantile([0.25,0.5,0.75])

print(Q)
```

```python
#todas as dancabilidades
fig = plt.figure(figsize=(40,6))

ax = sns.boxplot(x="gênero", y="dançabilidade", data=spotify)
ax.set_xticklabels(ax.get_xticklabels(),rotation=30)
plt.plot()
```

```python
#apenas subgrupos
spotify_subgrupo = pd.DataFrame(
    spotify[(
        (spotify['dançabilidade'] >=  20) 
        & (spotify['dançabilidade'] <= 40)
    )]
)

fig = plt.figure(figsize=(20,6))

ax = sns.boxplot(x="gênero", y="dançabilidade", data=spotify_subgrupo)
ax.set_xticklabels(ax.get_xticklabels(),rotation=30)
plt.plot()
```

```python
#agrupar
info_sobre_gen = spotify.groupby('gênero').mean()

info_sobre_gen.head()
```

```python
#apenas artistas selecionados
spotify_subgrupo2 = spotify.loc[(spotify['artista'] == 'Train') 
              | (spotify['artista'] == 'Eminem') 
              | (spotify['artista'] == 'Kesha')
              | (spotify['artista'] == 'Lady Gaga') 
              | (spotify['artista'] == 'Bruno Mars')
              | (spotify['artista'] == 'Mark Ronson')
              | (spotify['artista'] == 'DJ Khaled')
              | (spotify['artista'] == 'Ed Sheeran')
              ]

display(spotify_subgrupo)
```

```python
#quantas vezes Lady Gaga apareceu nos top 10 durante os periodos

spotify_subgrupo2['artista'].value_counts()
```

```python
#gráfico de pizza

spotify_subgrupo2['artista'].value_counts().plot(kind='pie', title='Aparição dos artistas nos top 10', figsize=(10,8) )
```

Há correlação entre dançabilidade e ritmo?

```python
plt.figure
sns.heatmap(spotify.corr(), linewidths=3.1, annot=True, center=1)
```

```python
#outra forma de visualização

sns.relplot(x='positividade', y='dançabilidade', data=spotify)
```

```python
sns.relplot(x='acústica', y='energia', data=spotify);
plt.show()
```

```python
sns.relplot(x='acústica', y='oralidade', data=spotify);
plt.show()
```

```python
plt.figure(figsize=(50,20))
sns.violinplot(x='ano', y='dançabilidade', data=spotify, scale='count', estimator='mean')
plt.xlabel('ano', fontsize=30)
plt.ylabel('dançabilidade', fontsize=30)
plt.show()
```
