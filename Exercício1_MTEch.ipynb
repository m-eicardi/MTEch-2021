###1. Perguntas 

*1.1. Quais são as 5 Disciplines (esportes) que mais premiaram medalhas?*

```python
dados['Discipline'].unique()
```

```python
top5_esportes = pd.DataFrame(dados['Discipline'].value_counts())

top5_esportes.head()
```

Os 5 esportes que mais premiaram medalhas são: Atletismo, Remo, Natação, Ginástica Artística e Esgrima.

*1.2. Quais são os 10 atletas que mais ganharam medalhas?*


```python
top10_atletas = pd.DataFrame(dados['Athlete'].value_counts())

top10_atletas.head(10)
```

Os 10 atletas que mais ganharam medalhas foram: LATYNINA, PHELPS, ANDRIANOV, SHAKHLIN, ONO, MANGIROTTI, FISCHER, THOMPSON, TORRES E NEMOV.

*1.3. Quantas medalhas foram premiadas desde 1896?*

```python
dados['Medal'].unique()
```

```python
#número de cada medalha
dados['Medal'].value_counts()
```

```python
#total de medalhas premiadas
dados['Medal'].value_counts().sum()
```

O total de medalhas premiadas desde 1896 é igual a 29.216
